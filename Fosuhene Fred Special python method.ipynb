2. comparing string "str" and developer representation "repr"

```python
class Student:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Name: {self.name}"

    def __repr__(self):
        return f"Student('{self.name}')"

s = Student("Fred")
print(s)
print(repr(s))
```

The length method

```python
class Playlist:
    def __init__(self, songs):
        self.songs = songs

    def __len__(self):
        return len(self.songs)

p = Playlist(["Song1", "Song2", "Song3"])
print(len(p))
```

Equality comparison "__eq__""

```python
class Student:
    def __init__(self, student_id):
        self.student_id = student_id

    def __eq__(self, other):
        return self.student_id == other.student_id

s1 = Student(101)
s2 = Student(101)
print(s1 == s2)
```

The Addition

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)
print(v1 + v2)
```

The subtraction "__sub__()"

```python
class Vector:
    def __init__(self, x):
        self.x = x

    def __sub__(self, other):
        return Vector(self.x - other.x)

    def __str__(self):
        return str(self.x)

v1 = Vector(10)
v2 = Vector(3)
print(v1 - v2)
# Output: 7
```

The Multiplication "__mul__()"

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __mul__(self, other):
        return Number(self.value * other.value)

    def __str__(self):
        return str(self.value)

n1 = Number(5)
n2 = Number(4)
print(n1 * n2)
# Output: 20
```

Item Access "__getitem__()"

```python
class Grades:
    def __init__(self):
        self.data = [85, 90, 78]

    def __getitem__(self, index):
        return self.data[index]

g = Grades()
print(g[0])
print(g[1])
```

Item Assignment "__setitem__()"

```python
class Grades:
    def __init__(self):
        self.data = [85, 90, 78]

    def __setitem__(self, index, value):
        self.data[index] = value

g = Grades()
g[1] = 100
print(g.data)
# Output: [85, 100, 78]
```

Testing for Membership "__contains__()"

```python
class Course:
    def __init__(self):
        self.students = ["Ella", "Sandra", "Joe"]

    def __contains__(self, student):
        return student in self.students

c = Course()
print("Joe" in c)
print("Gifty" in c)
# Output:
# True
# False
```

Callable objects "__call__()"

```python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, number):
        return number * self.factor

double = Multiplier(2)
print(double(10))
# Output: 20
```

Iteration support "__iter__(), __next__()"

```python
class Counter:
    def __init__(self, limit):
        self.limit = limit
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.limit:
            raise StopIteration
        value = self.current
        self.current += 1
        return value

for num in Counter(5):
    print(num)
```

Boolean Evaluation "__bool__()"

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def __bool__(self):
        return self.balance > 0

account = BankAccount(500)
if account:
    print("Account is active")
# Output: Account is active
```
